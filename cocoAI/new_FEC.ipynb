```python
from cocoAI.FEC import load_excel_data, load_nomenclature


excel_path_list = list(DATA_PATH.glob("202*xls*"))[:2]
df = load_excel_data(excel_path_list)
df = df.reset_index()
dfnom = load_nomenclature()
dfnom = dfnom.reset_index()

# main(excel_path_list)
```

```python
df['Classe'] = df['Compte'].apply(lambda x:str(x)[:1])
df['niveau1'] = df['Compte'].apply(lambda x:str(x)[:2])
df['niveau2'] = df['Compte'].apply(lambda x:str(x)[:3])
df['niveau3'] = df['Compte'].apply(lambda x:str(x)[:4])
df['niveau4'] = df['Compte'].apply(lambda x:str(x)[:5])
dft = df.merge(dfnom,on='niveau1')
```
