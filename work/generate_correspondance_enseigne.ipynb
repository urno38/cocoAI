```python
import numpy as np
import re


def return_ville_adresse(x):
    try:
        if "-" not in x:
            return {"adresse": np.nan, "ville": np.nan, "CP": np.nan}
        else:
            if len(x.split("-")) == 1:
                return {"adresse": np.nan, "ville": np.nan, "CP": np.nan}
            elif len(x.split("-")) == 2:
                for el in x.split("-"):
                    if re.match(r"^\d{5} ", el.strip()):
                        try:
                            ville = el.strip().split()[1].strip()
                            CP = el.strip().split()[0]
                            return {"adresse": np.nan, "ville": ville.strip(), "CP": CP}
                        except:
                            return {"adresse": np.nan, "ville": np.nan, "CP": np.nan}
                    else:
                        return {"adresse": x.strip(), "ville": np.nan, "CP": np.nan}
            elif len(x.split("-")) == 3:
                for el in x.split("-"):
                    if re.match(r"^\d{5}", el.strip()):
                        try:
                            ville = el.strip().split()[1].strip()
                            CP = el.strip().split()[0]
                            return {"adresse": np.nan, "ville": ville.strip(), "CP": CP}
                        except:
                            return {"adresse": np.nan, "ville": np.nan, "CP": np.nan}
                    else:
                        adresse = el.strip()
                        # ville = np.nan
                        # CP = np.nan
                return {"adresse": adresse, "ville": ville, "CP": CP}
    except:
        pass
    return {"adresse": np.nan, "ville": np.nan, "CP": np.nan}


name = [p.name for p in li if re.match(r"^[a-zA-z]", p.name)]
df = pd.DataFrame(name, columns=["folder"])
df["ENSEIGNE"] = df["folder"].apply(
    lambda x: (
        x.split("-")[0].strip().replace("(Le)", "").replace("(Les)", "").replace("(Le", "").replace("(La)", "").strip()
        if x.split("-")
        else np.nan
    )
)

for i in df["folder"].values:
    print(i)
    print(return_ville_adresse(i))

df["ville"] = df["folder"].apply(lambda x: return_ville_adresse(x)["ville"])
df["adresse"] = df["folder"].apply(lambda x: return_ville_adresse(x)["adresse"])
df["CP"] = df["folder"].apply(lambda x: return_ville_adresse(x)["CP"])
```
