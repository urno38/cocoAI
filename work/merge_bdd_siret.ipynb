```python
import json
import numpy as np
from cocoAI.folder_tree import get_df_folder_possibles
from common.identifiers import get_etablissement_name
from common.path import WORK_PATH


df = get_df_folder_possibles()
# folder_possibles = df[~df.siret.isna()].folder.values
print(df['siret'])
t = []
for k,i in df.set_index('folder')['siret'].items():
    print(k,i)
    try:
        get_etablissement_name(i)
    except:
        t.append(k)

def load_json_file(path):
    with open(path, "r", encoding="utf-8") as file:
        di = json.load(file)
    return di

di = load_json_file(WORK_PATH / "siret_databank.json")
new_di = {k:v for k,v in di.items() if k not in t}  

def write_json_file(di, path):
    with open(path, "w", encoding="utf-8") as f:
        f.write(json.dumps(di, ensure_ascii=False))
    return
write_json_file(new_di,WORK_PATH / "siret_databank2.json")
```
